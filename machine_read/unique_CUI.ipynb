```python
import pandas as pd
```

```python
reach = pd.read_csv('staging_mr_greentea_202108132247.csv')
```

```python
reach.head()
```

```python
df1 = pd.DataFrame()
df2 = pd.DataFrame()
```

```python
df1['name'] = reach.subject_name
df1['cui'] = reach.subject_cui
df2['name'] = reach.object_name
df2['cui'] = reach.object_cui
```

```python
df1 = df1.append(df2)
df1 = df1.drop_duplicates()
```

```python
df1 = df1.reset_index()
```

```python
df1 = df1.drop(['index'], axis=1)
df1
```

```python
df1.to_csv('unique_umls_name_cui.tsv', sep='\t', index=False)
```
