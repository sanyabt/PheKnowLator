```python
import pandas as pd
import numpy as np
```

```python
reach = pd.read_csv('greentea_pmid_all_predicates_umls.tsv', sep='\t')
df_new = pd.DataFrame()
```

```python
reach.head()
```

```python
df_new['subject_name'] = reach.subject_name
df_new['subject_cui'] = reach.subject_cui
df_new['subject_source'] = reach.subject_source
df_new.info()
```

```python
df2 = pd.DataFrame()
df2['subject_name'] = reach.object_name
df2['subject_cui'] = reach.object_cui
df2['subject_source'] = reach.object_source
df_new = df_new.append(df2)
```

```python
df_new.info()
```

```python
df_new = df_new.drop_duplicates()
df_new.info()
```

```python
df_new = df_new[pd.isna(df_new['subject_cui']) == False]
df_new.info()
```

```python
df_new.head()
```

```python
result = {}
```

```python
def add_to_dict(row):
    result[row['subject_source']] = row['subject_cui']
```

```python
df_new.apply(add_to_dict, axis=1)
```

```python
len(result)
```

```python
result['Tea']
```

```python
def get_cui(row):
    name= row['agent_name']
    if name in result:
        return result[name]
    else:
        return 0
```

```python
df = pd.DataFrame()
df['agent_name'] = reach.subject_source
df['grounding'] = reach.subj_reach_grounding
df['mapping'] = reach.subj_map_reach
```

```python
dfo = pd.DataFrame()
dfo['subject_name'] = reach.object_source
dfo['subject_cui'] = reach.obj_reach_grounding
dfo['subject_source'] = reach.subj_map_reach
df = df.append(dfo)
```

```python
df.info()
```

```python
df['cui'] = df.apply(get_cui, axis=1)
```

```python
save_df = pd.DataFrame()
save_df['agent_name'] = df['agent_name']
save_df['grounding'] = df['grounding']
save_df['other_mapping'] = df['mapping']
save_df['cui'] = df['cui']
save_df.info()
```
